### Подготовка моделей

```python
df.info()
```

```python
features = df.drop('price',axis = 1)
target = df['price']
```

##### dummies

```python
features1 = features.copy()
features1.sample(3)
```

```python
features1 = pd.get_dummies(data=features,columns =obj_col, drop_first= True, prefix = obj_col)
```

```python
features1
```

```python
features1.reset_index(inplace=True, drop=True)
target.reset_index(inplace=True, drop=True)
print(features1['vehicletype_convertible'].count())
```

```python
features1.info()   
```

##### Стандартизация

```python
features_train, features_common, target_train, target_common = train_test_split(
   features1, target, test_size=0.20, random_state=12345)
features_valid, features_test, target_valid, target_test = train_test_split(
   features_common, target_common, test_size=0.50, random_state=12345)
```

```python
features_train
target_train
```

```python
# теперь трансформируем скейлером только после пормирования всего набора фичей
# только фичи без признаков

ct = ColumnTransformer([
        ('unknown', StandardScaler(), ['registrationyear', 'power', 'kilometer', 'registrationmonth','postalcode'])
    ], remainder='passthrough')

features_train=ct.fit_transform(features_train)
features_test=ct.transform(features_test)
features_valid=ct.transform(features_valid)
```

```python
print(features_train.shape)
print(features_test.shape)
print(features_valid.shape)
print(features_train)
print(features_test)
print(features_valid)
```

### Обучение моделей

##### LinearRegression()

```python
%%time
import numpy as np
from sklearn.linear_model import LinearRegression
X = features_train
y =target_train
reg = LinearRegression().fit(features_train,target_train)
#reg.score(features_valid,target_valid)
```

```python
%%time
y_pred = reg.predict(features_valid)
y_true = target_valid

#print('r2_score:',r2_score(y_true, y_pred))
print('RMSE:',mean_squared_error(y_true, y_pred, squared = False))
```

##### RandomForestRegressor

```python
%%time

regr = RandomForestRegressor(max_depth=5, random_state=123456,n_jobs=-1,n_estimators=150,min_samples_split = 15)
regr.fit(features_train,target_train)
```

```python
%%time
predict = regr.predict(features_valid)
y_true = target_valid
print('r2_score:',r2_score(y_true, predict))
print('RMSE:',mean_squared_error(y_true, predict, squared = False))
```

#####   + LightGBM

```python
%%time
# defining parameters 
params = {
    'task': 'train', 
    'boosting': 'gbdt',
    'objective': 'regression',
    'num_leaves': 10,
    'learnig_rate': 0.05,
    'metric': {'l2','l1'},
    'verbose': -1
}
# laoding data
y_pred = reg.predict(features_valid)
y_true = target_valid

lgb_train = lgb.Dataset(features_train, target_train)
lgb_eval = lgb.Dataset(features_valid, target_valid, reference=lgb_train)
# fitting the model
model = lgb.train(params,
                 train_set=lgb_train,
                 valid_sets=lgb_eval,
                 early_stopping_rounds=30)
# prediction
y_pred = model.predict(features_valid)

```

```python
# accuracy check
rmse = mean_squared_error(y_pred, target_valid, squared = False)
#rmse = mse**(0.5)
#print("MSE: %.2f" % mse)
print("RMSE: %.2f" % rmse) 
```
